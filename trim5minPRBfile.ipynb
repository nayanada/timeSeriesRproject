```python

def trim5minPRBfile(filename, outfile):
    # type: (object, object) -> object
    xls = pd.ExcelFile(filename)
    df = pd.read_excel(xls, sheetname=0, skiprows=[0, 1])
    col = pd.Series(df.columns.values)
    col[0] = 'complexinfo'
    col = col[:4].append(col.str.extract(r'(.*)(\(.*\))')[0][4:])
    df.columns = col
    # make new columns...
    strSeries = df['complexinfo']
    delLines = strSeries.str.contains(u'[총소] 계')
    df = df[~delLines]
    df['eNB'] = strSeries.str.extract(r'.*(\d-\d{5}).*(\d{2}:\d{2})', expand=False)[0]
    df['time'] = strSeries.str.extract(r'.*(\d-\d{5}).*(\d{2}:\d{2})', expand=False)[1]
    df.to_csv(outfile, encoding="euc-kr")  # hangeul is not readable with utf-8.

#######################################
import fnmatch
fileList = fnmatch.filter(os.listdir('.'), 'DLUL*5M*.xls')

for file in fileList:
    fdate = re.search('2016(\d{4})', file, flags=0).group(1)
    trim5minPRBfile(file, "pss5min0{0}.csv".format(fdate))

#################################
```
